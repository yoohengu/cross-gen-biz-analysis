향후 가능 분석
구독 여부(FN)와 연령대(age)의 상관관계를 분석하여 타겟 마케팅 전략 수립 가능

```python
#  구독하는 사람(1)과 안 하는 사람(0)의 평균 나이는 다를까?
print(df.groupby('FN')['age'].mean())
# 구독하는 사람(1)과 안 하는 사람(0)의 평균 나이는 크게 다르지 않다
```

```python
# 나이대를 10대, 20대... 순으로 나누기
df['age_group'] = (df['age'] // 10) * 10

# 나이대별로 패션 뉴스 구독 비율 확인
fn_rate_by_age = df.groupby('age_group')['FN'].mean()
print(fn_rate_by_age)
```
