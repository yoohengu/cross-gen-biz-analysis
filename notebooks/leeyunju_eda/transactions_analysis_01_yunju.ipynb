```python
# 1. 현재 데이터의 나이별 개수를 셉니다.
age_counts = customers['age'].value_counts()

# 2. 데이터에 존재하는 최소 나이부터 최대 나이까지 빈틈없는 범위를 만듭니다.
# (예: 16세부터 99세까지 1살 단위로 생성)
start_age = int(customers['age'].min())
end_age = int(customers['age'].max())
all_ages = range(start_age, end_age + 1)

# 3. 모든 나이를 순회하며 개수를 출력합니다.
print(f"### {start_age}세부터 {end_age}세까지 상세 인원수 ###")
for age in all_ages:
    # 해당 나이의 데이터가 있으면 개수를 가져오고, 없으면 0을 가져옵니다.
    count = age_counts.get(age, 0)
    print(f"{age}세: {int(count):,}명")
```

```python
#모든 값이 일치한 중복된 행 확인_drop 할지 말지 결정해야할듯
transactions.duplicated().sum()
```

```python
#데이터프레임 복사 원본 데이터 보존을 위해서 copy() 매서드 사용
df_tran = transactions.copy()
```

```python
#데이터프레임 복사 원본 데이터 보존을 위해서 copy() 매서드 사용
df_cust = customers.copy()
```

```python
#데이터프레임 복사 원본 데이터 보존을 위해서 copy() 매서드 사용 
#모든 테이블을 복사해서 사용해볼 것임. 
df_art = articles.copy()
```

```python
#customer_id 열의 데이터 타입을 문자열로 변환
df_tran['customer_id']= df_tran['customer_id'].astype(str)
```

```python
#customer_id 열의 데이터 타입을 문자열로 변환_조인을 위해 동일한 타입으로 맞춰줌. 은솔님이 조언해주심.
df_cust['customer_id']= df_cust['customer_id'].astype(str)
```

```python
#article_id 열의 데이터 타입을 문자열로 변환
df_tran['article_id']= df_tran['article_id'].astype(str)
```

```python
#article_id 열의 데이터 타입을 문자열로 변환
df_art['article_id'] = df_art['article_id'].astype(str)
```

```python
#null 값 과 Dtype 확인
df_tran.info()
```

```python
#t_dat 열의 데이터 타입을 datetime 형식으로 변환
df_tran['t_dat'] = pd.to_datetime(df_tran['t_dat'], format='%Y-%m-%d')
```

```python
#datetime 형식 변환 확인
df_tran.info()
```

```python
#현재 분석 중인 H&M 데이터셋의 가격(price)은 대부분 0.016932 같은 매우 작은 소수점 형태입니다. 
#여기서 바로 int로 바꿔버리면 거의 모든 가격이 0이 되어버리는 대참사가 발생할 수 있어요!
#만약 숫자를 보기 편하게 만드는 것이 목적이라면, 아래처럼 특정 값을 곱한 뒤에 변환하는 것을 추천합니다.
#price 열의 데이터가 float 형식이므로 어떻게 처리하는게 가독성에 좋은지 고민 하다가 제미나이를 통해 조언을 얻음.
```

```python
#100을 곱한 것보다 조금 더 큰 1000을 곱해서 가격을 SEK 단위로 변환_컬럼명을 새로 만듦.
df_tran['price_sek'] = (df_tran['price'] * 1000).astype(int)
```

```python
#int 형식 변환 확인
df_tran.info()
```

```python
df_tran
```

```python
#소수점 둘째 자리까지 반올림한 가격 열 생성했어야 했는데 잘못해서 다시 하고 컬럼을 새로 만듦.
df_tran['price_sek'] = (df_tran['price'] * 1000).round(2)
```
